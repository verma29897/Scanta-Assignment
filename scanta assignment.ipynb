Merging All Data Sets

```python
final_data=df1.merge(df2.merge(df3.merge(df4,on='time'),on='time'))
```

Calculate power_value=voltag_value * current_value

```python

final_data['power_value']=final_data['voltage_value'].mul(final_data['current_value'],axis=0)
```

```python
display(final_data)
```

```python
#ploting voltage_value and current_value
sns.lineplot(x=final_data['voltage_value'],y=final_data['current_value'])
```

```python
#ploting voltage_value and power_value
sns.lineplot(x=final_data['voltage_value'],y=final_data['power_value'])
```

```python
#ploting power_value and current_value
sns.lineplot(x=final_data['power_value'],y=final_data['current_value'])
```

```python
delta_time=[]
for i in range(1,13510):
    delta_time.append((final_data.loc[i,'time']-final_data.loc[i-1,'time']))
delta_time.insert(0,0)
delta_time=pd.Series(delta_time)
final_data['delta_time']=delta_time
    
    
    
```

```python
display(final_data)
```

```python
display(final_data.tail())
```

```python
final_data['delta_energy']=final_data['power_value'].mul(final_data['delta_time'],axis=0)
```
